Usando el dataframe de los albumes que hemos creado en el ejercicio anterior. ¿Cuál es la media y la desviación estándar de "danceability", "acousticness" y "speechiness" de cada artista?. Muesrtra los resultados en tres dataframes diferentes, uno para cada métrica ("danceability", "acousticness" y "speechiness"). Ordena los resultados de mayor a menor en base a la media.

```python
album_dance_media=df_album.groupby('artist')['danceability'].mean().reset_index()
album_dance_media
```

```python
album_dance_desv=df_album.groupby('artist')['danceability'].std().reset_index()
```

```python
df_dance=album_dance_media.merge(album_dance_desv, how='inner', on='artist', suffixes= ('media','desviacion'))

df_dance.head(3)
```

```python
album_acous_media=df_album.groupby('artist')['acousticness'].mean().reset_index()
album_acous_desviacion=df_album.groupby('artist')['acousticness'].std().reset_index()
```

```python
df_acus=album_acous_media.merge(album_acous_desviacion, how='inner', on='artist', suffixes= ('media','desviacion'))

df_acus.sample(3)
```

```python
album_spe_media=df_album.groupby('artist')['speechiness'].mean().reset_index()
album_spe_desviacion=df_album.groupby('artist')['speechiness'].std().reset_index()
```

```python
df_spe=album_spe_media.merge(album_spe_desviacion, how='inner', on='artist', suffixes= ('media','desviacion'))

df_spe.sample(5)
```
